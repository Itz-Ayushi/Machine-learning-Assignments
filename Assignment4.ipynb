```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Initialize the Chrome driver
driver = webdriver.Chrome()

try:
    # Navigate to the website
    driver.get("https://quotes.toscrape.com/js/")
    
    # Wait for the page to load completely (the quotes to appear)
    # This is crucial for JavaScript-rendered pages
    wait = WebDriverWait(driver, 10)
    wait.until(EC.presence_of_element_located((By.CLASS_NAME, "quote")))
    
    # Optional: add a small delay to ensure everything is loaded
    time.sleep(2)
    
    # Now find the quotes
    quotes = driver.find_elements(By.CLASS_NAME, "text")
    
    # Print each quote
    for q in quotes:
        print(q.text)

finally:
    # Always quit the driver to release resources
    driver.quit()
```

```python
# Import necessary libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
import chromedriver_autoinstaller
import pandas as pd
import time

# Automatically install the correct ChromeDriver version
chromedriver_autoinstaller.install()

# Configure Chrome (headless mode optional)
options = webdriver.ChromeOptions()
options.add_argument("--headless")      # comment out this line if you want to see the browser
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Start the browser
driver = webdriver.Chrome(options=options)

# Open IMDb Top 250 page
url = "https://www.imdb.com/chart/top/"
driver.get(url)

# Wait a bit to make sure page fully loads
time.sleep(3)

# Extract movie details
movies = driver.find_elements(By.CSS_SELECTOR, "li.ipc-metadata-list-summary-item")

data = []
rank = 1

for movie in movies:
    try:
        title = movie.find_element(By.CSS_SELECTOR, "h3.ipc-title__text").text
        # The title often looks like "1. The Shawshank Redemption"
        title_clean = title.split(". ")[1] if ". " in title else title

        year = movie.find_element(By.CSS_SELECTOR, "span.cli-title-metadata-item").text
        rating = movie.find_element(By.CSS_SELECTOR, "span.ipc-rating-star--rating").text

        data.append({
            "Rank": rank,
            "Movie Title": title_clean,
            "Year of Release": year,
            "IMDB Rating": rating
        })
        rank += 1
    except Exception as e:
        # Skip any malformed entries
        continue

# Close the browser
driver.quit()

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Display first few rows
print(df.head())

# Export to CSV
df.to_csv("imdb_top250.csv", index=False, encoding="utf-8")

print("\n✅ Data saved successfully to imdb_top250.csv")
```

```python
# Import necessary libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
import chromedriver_autoinstaller
import pandas as pd
import time

# Automatically install the correct ChromeDriver version
chromedriver_autoinstaller.install()

# Setup Chrome options
options = webdriver.ChromeOptions()
options.add_argument("--headless")     # Run in background
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Launch browser
driver = webdriver.Chrome(options=options)

# Open Time and Date Weather page
url = "https://www.timeanddate.com/weather/"
driver.get(url)
time.sleep(3)  # Wait for page to load

# Find all country links listed under "World Weather"
countries = driver.find_elements(By.CSS_SELECTOR, "section#wf-map a")

city_data = []

# Limit scraping to top 10 countries (optional, to avoid long runtime)
for country in countries[:10]:
    country_url = country.get_attribute("href")
    driver.get(country_url)
    time.sleep(2)

    # Extract the cities table
    try:
        rows = driver.find_elements(By.CSS_SELECTOR, "table.zebra.tb-wt tbody tr")

        for row in rows:
            try:
                city_name = row.find_element(By.CSS_SELECTOR, "td:nth-child(1) a").text
                temperature = row.find_element(By.CSS_SELECTOR, "td:nth-child(2)").text
                condition = row.find_element(By.CSS_SELECTOR, "td:nth-child(3)").text

                city_data.append({
                    "City Name": city_name,
                    "Temperature": temperature,
                    "Weather Condition": condition
                })
            except Exception:
                continue
    except Exception:
        continue

# Close browser
driver.quit()

# Create DataFrame
df = pd.DataFrame(city_data)

# Show first few rows
print(df.head())

# Export to CSV
df.to_csv("weather.csv", index=False, encoding="utf-8")

print("\n✅ Weather data saved successfully to weather.csv")
```
