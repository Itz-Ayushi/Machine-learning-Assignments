```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.decomposition import PCA
```

```python
cols = ["symboling", "normalized_losses", "make", "fuel_type", "aspiration",
        "num_doors", "body_style", "drive_wheels", "engine_location",
        "wheel_base", "length", "width", "height", "curb_weight",
        "engine_type", "num_cylinders", "engine_size", "fuel_system",
        "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm",
        "city_mpg", "highway_mpg", "price"]

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, names=cols, na_values="?")
```

```python
df.head()
```

```python
for col in df.columns:
    if df[col].dtype == "object":
        df[col].fillna(df[col].mode()[0], inplace=True)  # categorical → mode
    else:
        df[col].fillna(df[col].mean(), inplace=True)    

df=df.dropna(subset=["price"])

df
```

```python
num_map = {"two": 2, "three": 3, "four": 4, "five": 5, "six": 6, "eight": 8, "twelve": 12}
df["num_doors"] = df["num_doors"].replace(num_map).astype(int)
df["num_cylinders"] = df["num_cylinders"].replace(num_map).astype(int)
```

```python
df = pd.get_dummies(df, columns=["body_style", "drive_wheels"], drop_first=True)
```

```python
label_cols = ["make", "aspiration", "engine_location", "fuel_type"]
le = LabelEncoder()
for col in label_cols:
    df[col] = le.fit_transform(df[col])
```

```python
df["fuel_system"] = df["fuel_system"].apply(lambda x: 1 if "pfi" in x else 0)

df["engine_type"] = df["engine_type"].apply(lambda x: 1 if "ohc" in x else 0)
```

```python
X = df.drop("price", axis=1)
y = df["price"].astype(float)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
print("Linear Regression Results:")
print("R² Score:", r2_score(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
pca = PCA(n_components=15)  # keep 15 components
X_pca = pca.fit_transform(X_scaled)

Xp_train, Xp_test, yp_train, yp_test = train_test_split(X_pca, y, test_size=0.3, random_state=42)

lr_pca = LinearRegression()
lr_pca.fit(Xp_train, yp_train)

yp_pred = lr_pca.predict(Xp_test)
print("\nPCA + Linear Regression Results:")
print("R² Score:", r2_score(yp_test, yp_pred))
print("RMSE:", np.sqrt(mean_squared_error(yp_test, yp_pred)))
```
